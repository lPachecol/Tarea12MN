```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del ejercicio 3
t_euler = np.array([0, 0.5, 1.0, 1.5, 2.0])
y_euler = np.array([1, 1.5, 2.25, 3.375, 5.0625])  # Ejemplo con y' = y, h = 0.5

# Punto a interpolar: y(0.25) y y(0.93)
def interpolar_lineal(x0, x1, y0, y1, x):
    return y0 + ((y1 - y0) / (x1 - x0)) * (x - x0)

y_025 = interpolar_lineal(0, 0.5, 1, 1.5, 0.25)
y_093 = interpolar_lineal(0.5, 1.0, 1.5, 2.25, 0.93)

# Gráfica
t_vals = np.linspace(0, 2, 100)
y_vals = np.exp(t_vals)

plt.plot(t_vals, y_vals, label="Solución exacta")
plt.plot(t_euler, y_euler, 'o-', label="Euler")
plt.plot([0.25, 0.93], [y_025, y_093], 'rx', label="Interpolación")
plt.title("Interpolación lineal - Literal a")
plt.legend()
plt.grid()
plt.show()

```

```python
# Interpolar en y(1.25) y y(1.93)
y_125 = interpolar_lineal(1.0, 1.5, 2.25, 3.375, 1.25)
y_193 = interpolar_lineal(1.5, 2.0, 3.375, 5.0625, 1.93)

# Gráfica
plt.plot(t_vals, y_vals, label="Solución exacta")
plt.plot(t_euler, y_euler, 'o-', label="Euler")
plt.plot([1.25, 1.93], [y_125, y_193], 'rx', label="Interpolación")
plt.title("Interpolación lineal - Literal b")
plt.legend()
plt.grid()
plt.show()
```

```python
# Interpolar en y(2.10) y y(2.75)
# Para esto necesitamos extender los datos de Euler (solo a modo demostrativo, lo ideal es tener más pasos)
t_euler_ext = np.array([1.5, 2.0, 2.5, 3.0])
y_euler_ext = np.array([3.375, 5.0625, 7.5938, 11.3906])

y_210 = interpolar_lineal(2.0, 2.5, 5.0625, 7.5938, 2.10)
y_275 = interpolar_lineal(2.5, 3.0, 7.5938, 11.3906, 2.75)

# Gráfica
plt.plot(t_vals, y_vals, label="Solución exacta")
plt.plot(t_euler_ext, y_euler_ext, 'o-', label="Euler extendido")
plt.plot([2.10, 2.75], [y_210, y_275], 'rx', label="Interpolación")
plt.title("Interpolación lineal - Literal c")
plt.legend()
plt.grid()
plt.show()
```
