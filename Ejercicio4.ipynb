```python
import matplotlib.pyplot as plt
import numpy as np

def y_real(t):
    return t / (1 + np.log(t))

def f(t, y):
    return (y / t) - (y / t)**2

t0, y0 = 1, 1
h = 0.1
t_final = 2

t_values = [t0]
y_values = [y0]
real_values = [y_real(t0)]
errors = [abs(y0 - y_real(t0))]

while t_values[-1] < t_final:
    t_n = t_values[-1]
    y_n = y_values[-1]
    y_next = y_n + h * f(t_n, y_n)
    t_next = t_n + h
    t_values.append(t_next)
    y_values.append(y_next)
    y_true = y_real(t_next)
    real_values.append(y_true)
    errors.append(abs(y_true - y_next))

# Imprimir resultados
print(" t\t y_aprox\t y_real\t\t error")
for t, y_a, y_r, err in zip(t_values, y_values, real_values, errors):
    print(f"{t:.2f}\t{y_a:.6f}\t{y_r:.6f}\t{err:.6f}")

# Gráfica
t_plot = np.linspace(t0, t_final, 100)
plt.plot(t_plot, y_real(t_plot), label="Solución real", linewidth=2)
plt.plot(t_values, y_values, 'o--', label="Euler", color='red')
plt.title("Ejercicio 4a: Comparación solución real vs Euler")
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
