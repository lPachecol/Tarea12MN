```python
import numpy as np
import matplotlib.pyplot as plt

# Derivadas para Taylor orden 2
def f(t, y):
    return t * np.exp(3*t) - 2*y

def df(t, y):
    return (1 + 3*t) * np.exp(3*t) - 2*f(t, y)

# Valores iniciales
t0, y0 = 0, 0
h = 0.5
t_values = [t0]
y_values = [y0]

# Método de Taylor orden 2
t = t0
y = y0
while t < 1:
    y = y + h * f(t, y) + (h**2 / 2) * df(t, y)
    t = t + h
    t_values.append(t)
    y_values.append(y)

# Gráfica
plt.plot(t_values, y_values, marker='o', label='Taylor orden 2')
plt.title("Ejercicio 6a - Taylor orden 2")
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Derivadas para Taylor orden 2
def f(t, y):
    return 1 + (t - y)**2

def df(t, y):
    return 2*(t - y)*(1 - f(t, y))

# Valores iniciales
t0, y0 = 2, 1
h = 0.5
t_values = [t0]
y_values = [y0]

# Método de Taylor orden 2
t = t0
y = y0
while t < 3:
    y = y + h * f(t, y) + (h**2 / 2) * df(t, y)
    t = t + h
    t_values.append(t)
    y_values.append(y)

# Gráfica
plt.plot(t_values, y_values, marker='o', label='Taylor orden 2')
plt.title("Ejercicio 6b - Taylor orden 2")
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
