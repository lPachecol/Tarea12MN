```python
import numpy as np
import matplotlib.pyplot as plt

# Derivadas hasta orden 3 para Taylor orden 4
def f(t, y):
    return t * np.exp(3*t) - 2*y

def f1(t, y):  # Primera derivada
    return (1 + 3*t) * np.exp(3*t) - 2*f(t, y)

def f2(t, y):  # Segunda derivada
    return (3 + 9*t) * np.exp(3*t) - 2*f1(t, y)

def f3(t, y):  # Tercera derivada
    return (9 + 27*t) * np.exp(3*t) - 2*f2(t, y)

# Parámetros iniciales
t0, y0 = 0, 0
h = 0.5
t_values = [t0]
y_values = [y0]

# Método de Taylor orden 4
t, y = t0, y0
while t < 1:
    y = y + h*f(t,y) + (h**2/2)*f1(t,y) + (h**3/6)*f2(t,y) + (h**4/24)*f3(t,y)
    t = round(t + h, 5)
    t_values.append(t)
    y_values.append(y)

# Gráfica
plt.plot(t_values, y_values, marker='o', label='Taylor orden 4')
plt.title("Ejercicio 7a - Taylor orden 4")
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Derivadas hasta orden 3 para Taylor orden 4
def f(t, y):
    return 1 + (t - y)**2

def f1(t, y):
    return 2*(t - y)*(1 - f(t, y))

def f2(t, y):
    return 2*(1 - f(t, y))**2 - 2*(t - y)*f1(t, y)

def f3(t, y):
    term1 = -2*f1(t, y)**2
    term2 = -2*(t - y)*f2(t, y)
    term3 = -4*(1 - f(t, y))*f1(t, y)
    return term1 + term2 + term3

# Parámetros iniciales
t0, y0 = 2, 1
h = 0.5
t_values = [t0]
y_values = [y0]

# Método de Taylor orden 4
t, y = t0, y0
while t < 3:
    y = y + h*f(t,y) + (h**2/2)*f1(t,y) + (h**3/6)*f2(t,y) + (h**4/24)*f3(t,y)
    t = round(t + h, 5)
    t_values.append(t)
    y_values.append(y)

# Gráfica
plt.plot(t_values, y_values, marker='o', label='Taylor orden 4')
plt.title("Ejercicio 7b - Taylor orden 4")
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
